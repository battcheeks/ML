<h3>Dataframe: A two-dimensional labeled data structure with columns of potentially different types. It is similar to excel table.</h3>

Can make data frame using dictionary, list of list

```python
data = {'Country': ['Belgium', 'India', 'Brazil'],
       'Capital': ['Brussels', 'New Delhi', 'Brasília'],
       'Population': [11190846, 1303171035, 207847528]}

df_sample = pd.DataFrame(data,
                         columns=['Country', 'Capital', 'Population'])
```

```python
df_sample
```

<h1>Boolean indexing</h1>
Returns rows where the stated condition returns true<br>

or -> condition 1 | condition 2 (or also works but throws ambiguity error for multiple conditions)<br>
and -> condition 1 & condition 2 (and also works but throws ambiguity error for multiple conditions<br>
not -> ~ (not condition)<br>
equal -> == Satisfying condition<br>
any() -> columns/rows with any value matching condition<br>
all() > columns/rows with all values matching some condition<br>

```python
df.head(100)
```

```python
df[(df.is_canceled == 1) | (df.iloc[:, 3] == 2015)].iloc[3:10]
```
